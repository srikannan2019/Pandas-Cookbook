```python
#memory usage savings by changing the data type
# though here in this example changing data type doesnt have impact this is an important step in data analysis
cols_new =['fractal_dimension_worst','smoothness_worst','texture_worst','diagnosis']
cols2 = df.loc[:,cols_new]

# to find the memory usage of the columns by memory _usage
original_memory= cols2.memory_usage(deep = True)
original_memory
```

```python
cols2.head()
cols2.dtypes

```

```python
#converting to a different data type using astype
#not all the columns could be converted to a data type .if a column has some missing values such as nanvalues ,it cant be converted to int dtype
#all missing values are of float64 dtype
#any column of integer dtype has a missing value will be changed ie is that column will be  of float value 
cols2['fractal_dimension_worst']=cols2['fractal_dimension_worst'].astype(np.int8)
```

```python
original_memory= cols2.memory_usage(deep = True)
original_memory
# the memory usage is reduced to 569 from 4552 for this column
```

```python
# to save even more memory ,the object dtype are converted to category dtype
#'object' dtype contains a mix of integers strings list tuples or timestamps in its colums 
#this is done by finding the number of unique values of columns of object dtype
#in this example only one column is object dtype ie diagnosis
df['diagnosis'].nunique()
```

```python
#convert this to category data
cols2['diagnosis'] = cols2['diagnosis'].astype('category')
original_memory= cols2.memory_usage(deep = True)
original_memory
```

```python
#Selecting smallest from the largest ie find the top 'n values' in a coulmn and find the bottom m values based on a different column
movie = pd.read_csv(r"C:\Users\SRIVIDHYA\Desktop\movie_metadata.csv")
movie.head(10).T
movie2 =movie[['imdb_score','budget','movie_title']]
# drop the missing values
movie2.dropna(axis= 0,how = 'any',inplace = True)
#inplace = True returns none and false  makes a copy of the object with the opeartionperformed.
movie2.head(5)
# find the movies that are in top 100 based on imdb score and with low budget(top 5)
#nlargest has a paramter keep that  takes first ,last,all.
#first keeps the first occurence of the duplicate rows
#last keeps the last occurence of the duplicate rows
#all keeps all rows
movie2.nlargest(100,'imdb_score').nsmallest(5,'budget').head()

#columns parameter with list of  column names takes the nlargest and nsmallest methods when there are duplicates values share the nth rank place in the first column in the list


```

```python
#Select the largest of each group by sorting

movie3 = movie[['movie_title','title_year','imdb_score']]
#sort values based on movie year and score put in a list by sort_values  method that sorts multiple columns in a time
movie4 = movie3.sort_values(['title_year','imdb_score'],ascending = False)
movie4.head()

    
```

```python
# drop the duplicates of each year just keeping the first row of each year
#drop_duplicates keeps the first occurence of unique row 

movie_to_year = movie4.drop_duplicates(subset='title_year')
movie_to_year.head(10)
```

```python
#to ascend one column and descend another column
movie5 = movie[['movie_title','title_year','content_rating','budget']]
#sort values based on movie year and score put in a list by sort_values  method that sorts multiple columns in a time,give the boolean values in ascending
movie6 = movie5.sort_values(['title_year','content_rating','budget'],ascending = [False,True,False])
movie6.head()
```

```python
# find the movies that are in top 100 based on imdb score and with low budget(top 5) but by using sort_values ...instead of nlargest and nsmallest
#movie2.nlargest(100,'imdb_score').nsmallest(5,'budget').head()
movie2.sort_values('imdb_score',ascending = False).head(100).sort_values('budget').head()

```
