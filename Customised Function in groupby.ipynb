```python
#to calculate the  % of suicides by country and and sex that are common to the age group
def pct_suicides_age(s,age_low,age_high):
    return s.between(age_low,age_high).mean()
df.groupby(['country','sex']) ['suicides_no'].agg(pct_suicides_age,10,50).head(5)
#in an agg function here the defined function is passed with low and high values 
#kwargs are keyword arguments collected in a dictionary 
#args are collected as tuple as non keyword arguments 

#the customised function when passed with pandas built in functions with args and kwargs raises an eror
#in that case use closures that have a function within a function 

   
```
