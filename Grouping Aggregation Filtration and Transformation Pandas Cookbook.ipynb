```python
#in any grouping function,find the grouping column,agg column and agg function
#groupby takes 4 functions as agg,filter,transform and apply
import pandas as pd
import numpy as np

df = pd.read_csv(r'C:\Users\SRIVIDHYA\Desktop\flights.csv')
df.shape
#2. to find the rows and columns
df.dtypes
#3.to find the data types
df.info()
#4. to find the number of non null values and data types
#split - apply - combine
#group by object - takes many inputs and gives one output
#aggregation has two categories as aggregation by columns and by functions such as sum,min.max,mean,count
#agg function creates multiple index on columns as aggregating columns on the toplevel and aggreagting functions on the bottom level 
df.head()
```

```python
#group by and aggregating single column and value
#groupby just creates the groupby object by validating the column for groupby only agg gives out the value
#place the groupby in the column and aggregate function as a dictionary pair

df.groupby('AIRLINE').agg({'ARRIVAL_DELAY':'mean'}).head()
#using numpy 
df.groupby('AIRLINE') ['ARRIVAL_DELAY'].agg(np.mean).head()
#also can be done as
#the direct way of groupby is for certain agg functions such as mean,min,max,median,su,count,std,var,size,describe,nunique,idxmin,idxmax

df.groupby(['AIRLINE'],as_index=False) ['ARRIVAL_DELAY'].mean().head()
#as_index = false in groupby gives the same behaviour as reset_index by removing multi index 
```

```python
#grouping multiple columns and aggre multiple colns and functions
#find the number of cancelled flights for every airline per weekday
#so the groupby column is by airline and weekday and agg coln is by cancelled flights and the agg fn is sum

df.groupby(['AIRLINE','DAY_OF_WEEK']) ['CANCELLED'] .sum().head(7)
```

```python
#find the number and percentage of cancelled and diverted flights for every airline every weekday
flight_df=df.groupby(['AIRLINE','DAY_OF_WEEK'])['CANCELLED','DIVERTED'].agg(['sum','mean']).head(7)
```

```python
#find the origin and destination for the total number of flights and the number and percentage of cancelled flights and avergae and variance of airtime
group_cols =['ORIGIN_AIRPORT','DESTINATION_AIRPORT']
agg_dict ={'CANCELLED':['sum','mean','size'],'AIR_TIME':['mean','var']}
flight_df1=df.groupby (group_cols).agg(agg_dict).head(7)
#size returns the total number of rows per group in an agg function
#count returns the number of non missing values per group
flight_df1


```

```python
#group by creates multi index on columns and rows
#multiindex with two levels 
#get_level_values method ---to make multi index to single index columns
level0 = flight_df1.columns.get_level_values(0)
level1 = flight_df1.columns.get_level_values(1)
flight_df1.columns = level0 + '-'+level1
flight_df1.head(7)
```

```python
#returns the row labels to a single level using reset_index
flight_df1.reset_index().head(7)
```
