```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import numpy as np

df = pd.read_csv(r'C:\Users\SRIVIDHYA\Desktop\train.csv')
df.head(5)
```

```python
pd.isnull(df).any()
#age,Cabin,Embarked  have null values 
```

```python
df.dropna()
#since the missing value for cabin the rows are dropped and are not taken for analysis
```

```python
#Filtering data according to requirements by conditions
df[df['Age'] > 30]
df[df['Sex'] == 'female']
femaleage = df[(df['Age'] > 30)& (df['Sex']== 'female')]
femaleage.describe()

##this shows that mean of women whose age  are  greater than 30 have 80 % chance  of survival ...
```

```python
df.groupby('Sex').Survived.value_counts()

#to find the chance of survival by sex 
#value_counts returns the count of values for unique values in the column
# shows male has higher chance of survival
```

```python
plt.plot(df["Survived"],df["Age"])
plt.xticks(rotation= 90)
plt.xlabel("Pclass")
plt.ylabel("Survived")
plt.title("People survived by passenger class list")
plt.show()
```
